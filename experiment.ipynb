```python
import ollama
```

```python
stream = ollama.chat(
    model='llama2',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)
```

```python

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```

```python
response = ollama.chat(model='llama2', messages=[
  {
    'role': 'user',
    'content': 'Why is the sky blue?',
  },
])
print(response['message']['content'])

```
